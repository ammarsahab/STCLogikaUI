Pisahkan tiap instance {id: n, qty: n} (jadi pembelian jenis produk beda di satu transaksi jadi kolom beda)

```python
nums2018_1=trx2018['purchase'].str.split('}, {', expand=True)
nums2019_1=trx2019['purchase'].str.split('}, {', expand=True)
nums2020_1=trx2020['purchase'].str.split('}, {', expand=True)
```

```python
nums2018_1['tid']=trx2018['tid']
nums2019_1['tid']=trx2019['tid']
nums2020_1['tid']=trx2020['tid']
```

```python
nums2018_1.index=nums2018_1.index+1
nums2019_1.index=nums2019_1.index+1
nums2020_1.index=nums2020_1.index+1
```

```python
nums2018_1.info()
nums2019_1.info()
nums2020_1.info()
```

wide to long reshaping: ada 12 kolom karena maks jenis barang yang dibeli di satu transaksi 12, dan pasti ada banyak NA. Jadi 12 kolom itu dibuat menjadi row, dan hapus na agar jml jenis produk per transaksi terlihat.

```python
nums2018_1=pd.melt(nums2018_1,id_vars=['tid'],value_vars=range(11))
nums2019_1=pd.melt(nums2019_1,id_vars=['tid'],value_vars=range(12))
nums2020_1=pd.melt(nums2020_1,id_vars=['tid'],value_vars=range(9))
```

lalu, split id dan qty di tiap row dan extract angka

```python
nums2018_1[['pid', 'qty']] = nums2018_1['value'].str.split(',', 1, expand=True)
nums2019_1[['pid', 'qty']] = nums2019_1['value'].str.split(',', 1, expand=True)
nums2020_1[['pid', 'qty']] = nums2020_1['value'].str.split(',', 1, expand=True)
```

```python
nums2018_1['pid']=nums2018_1['pid'].str.extract('(\d+)')
nums2018_1['qty']=nums2018_1['qty'].str.extract('(\d+)')
nums2019_1['pid']=nums2019_1['pid'].str.extract('(\d+)')
nums2019_1['qty']=nums2019_1['qty'].str.extract('(\d+)')
nums2020_1['pid']=nums2020_1['pid'].str.extract('(\d+)')
nums2020_1['qty']=nums2020_1['qty'].str.extract('(\d+)')
```

```python
nums2018_1['qty']=pd.to_numeric(nums2018_1['qty'])
nums2019_1['qty']=pd.to_numeric(nums2019_1['qty'])
nums2020_1['qty']=pd.to_numeric(nums2020_1['qty'])
```

Karena nums asli memiliki 11-12 kolom yang menandakan 11-12 maks produk yang berbeda yang dibeli di transakasi, tidak semua transaksi membeli produk sebanyak itu, maka dia akan jadi NA. Drop saja.

```python
nums2018_1=nums2018_1.dropna(subset=['pid'])
nums2019_1=nums2019_1.dropna(subset=['pid'])
nums2020_1=nums2020_1.dropna(subset=['pid'])
```

Kita asumsikan bahwa di purchase (sebagai representasi bon) suatu produk tidak tercatat double, jadi drop saja jika di transaksi tercatat produk double.

```python
nums2018_1 = nums2018_1.drop_duplicates(
  subset = ['tid', 'pid'],
  keep = 'last').reset_index(drop = True)
nums2019_1 = nums2019_1.drop_duplicates(
  subset = ['tid', 'pid'],
  keep = 'last').reset_index(drop = True)
nums2020_1 = nums2020_1.drop_duplicates(
  subset = ['tid', 'pid'],
  keep = 'last').reset_index(drop = True)
```
