```python
X = trains.loc[:,'prodvar':'spend']
y = trains.loc[:,'membership_class_trial':'membership_class_trial']
Xfalse = false.loc[:,'prodvar':'spend']
yfalse = false.loc[:,'membership_class_trial':'membership_class_trial']
y = y.astype(int)
yfalse = yfalse.astype(int)
```

```python
X = X.drop(['region'],axis=1)
X = X.drop(['catvar'],axis=1)
X = X.drop(['spendgrow'],axis=1)
X = X.drop(['qtygrow'],axis=1)
X = X.drop(['transgrow'],axis=1)
X = X.drop(['spend'],axis=1)

Xfalse = Xfalse.drop(['region'],axis=1)
Xfalse = Xfalse.drop(['catvar'],axis=1)
Xfalse = Xfalse.drop(['spendgrow'],axis=1)
Xfalse = Xfalse.drop(['qtygrow'],axis=1)
Xfalse = Xfalse.drop(['transgrow'],axis=1)
Xfalse = Xfalse.drop(['spend'],axis=1)
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = 123) 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
x_train = scaler.transform(X_train)
x_test = scaler.transform(X_test)
x_false = scaler.transform(Xfalse)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier = KNeighborsClassifier(n_neighbors=7)  
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
yf_pred = classifier.predict(x_false)
yf_pred
```

```python
classifier.predict_proba(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
print(confusion_matrix(yfalse, yf_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(classification_report(yfalse, yf_pred))
```
